```python
import sys
sys.path.append('..')

from src.datadict import DataDictionary
from src.template import Template
```

```python
# Lade DataDict
datadict = DataDictionary.from_csv("../data/datadict.csv")
print(f"✓ DataDict geladen: {len(datadict.fields)} Felder")
```

```python
# Erstelle Template
template = Template.from_datadict(datadict, name="Test Template from DataDict")
print(f"✓ Template erstellt mit {len(template.fields)} Feldern")
```

```python
# Zeige erste 3 Felder
print("Erste 3 Felder:")
for i, field in enumerate(template.fields[:3]):
    print(f"  {i+1}. {field.field_name} ({field.field_type})")
    print(f"      Label: {field.field_label}")
```

```python
# Speichere als YAML
template.to_yaml("../data/test_template.yaml")
print(f"✓ Template gespeichert zu: ../data/test_template.yaml")
```

```python
# Verifiziere: lade Template zurück
reloaded = Template.from_yaml("../data/test_template.yaml")
print(f"✓ Template reloaded: {len(reloaded.fields)} Felder")
print(f"  Name: {reloaded.name}")
```
